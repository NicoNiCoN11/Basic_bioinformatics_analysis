# 表格处理方法总结

##  插入&删除行或者列

```python
import pandas as pd
```

```python
## 读取列表
data_frame = pd.read_csv("GSE42872_IDS.csv", index_col= 0 )
```

```python
    data_frame.head(5)
```

```python
## 删除GSM316653列
data_frame = data_frame.drop(columns='GSM316653')
```

```python
data_frame.head(5)
```

```python
## 删除RFC2行
data_frame = data_frame.drop('RFC2', axis = 0)
```

```python
data_frame.head(5)
```

```python
## 删除指定数字的第几行第几列
## 删除后20列,以及前3行
data_frame = data_frame.iloc[3:, :58] # 左闭右开
data_frame.head(5)
```

## 将某几列或者某几行列表提取，建立新的列表

```python
## 读取文件
data_frame2 = pd.read_csv('GSE42872_cli.csv', index_col=0) 
data_frame2
```

```python
# 从数据框中提取第2列与第3列，作为新表格的两列
new_df = data_frame2.iloc[:,0:2]
new_df
```

## 将两个列表融合（进行ID转换）

```python
## 从data_frame2中获取title列
title_column = data_frame2["title"]
title_column
```

```python
# 将第一个DataFrame的列索引替换为"title"列的值
data_frame.columns = title_column
```
